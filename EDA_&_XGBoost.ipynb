## Informacin sobre las variables del dataset

### Datos del cliente bancario
1. age
2. job: tipo de trabajo
3. marital: estado civil (nota: "divorced" implica tanto divorciado como viudo)
4. education
5. default: no ha pagado alguna cuota de un crdito?

9. contact: tipo de comunicacin del ltimo contacto
10. day: da de la fecha de ltimo contacto
11. month: mes de la fecha de ltimo contacto
12. duration: duracin del ltimo contacto, en segundos

13. campaign: nmero de contactos llevados a cabo durante esta campaa y para este cliente (incluye el ltimo contacto)
14. pdays: nmero de dias que han pasado desde que el cliente fue contactado por ltima vez durante la pasada campaa (-1 significa que el cliente no fue contactado)
15. previous: nmero de contactos que se han llevado a cabo antes de esta campaa y para este cliente
16. poutcome: resultado de la ltima campaa de marketing

### Variable de salida (target):
17. y: tiene el cliente contratado un depsito a plazo fijo?

### housing vs loan
Lo primero que se me viene a la cabeza es saber si las variables `housing` y `loan` tienen una estrecha relacin o no. Ya que, si la persona tiene una hipoteca, es probable que tenga tambin un crdito/prstamo.Si es as, se podra eliminar una de ellas. Lo verificatemos con una tabla de contigencia.


```python
contingency_table = pd.crosstab(bank_df_pandas['housing'], bank_df_pandas['loan'])
sns.heatmap(contingency_table, annot=True, cmap='coolwarm', fmt='d')
plt.title('Tabla de Contingencia entre Housing y Loan')
plt.show()
```

```python
contingency_table.sum()
```

```python
# La suma por columna ser 100%
contingency_table_per = pd.crosstab(bank_df_pandas['housing'], bank_df_pandas['loan'], normalize='columns').round(3)
contingency_table_per
```

```python
# Realizando el test Chi-Square y evaluar independencia entre variables
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-Value: {p}")

if p < 0.05:
    print("S hay una asocicacin significativa entre housing y loan.")
else:
    print("No hay una asocicacin significativa entre housing y loan.")
```

# Handling Imbalanced Data

Como se pueden ver en el histogramas, pie charts, en los anlisis realizados, la data del target variable 'y' se encuentra desbalanceada. Es por ello que se proceder a realizar los siguentes procesamientos de data:

* **Oversamplig:** Para mejorar la representacin de las clases subrepresentadas. En este caso se eligir la variante SMOTENC del algoritmo SMOTE (Synthetic Minority Oversampling Technique) ya que permite trabajar con data mixta (con features continuos y categricos).

  * Recordemos que SMOTE genera data sinttica de la minora interpolando los datos existentes, es decir, generando datos similares a los ya existentes. A diferencia de ADASYN que genera data sinttica similar a los datos ms dificiles de aprender (cerca de las fronteras si lo vemos desde un punto de vista de KNN).
  * Se podra experimentar utilanzo ADASYN u algn otro algoritmo de Oversampling.

* **Undersamplig:** Reducir un poco las clases sobrerepresentadas. Es tentativo emplear el algoritmo TOME LINK; sin embargo, este algoritmo funciona solo para variables continuas, con lo cual habra primero que aplicar preprocesamiento a los features categricos para facilitar el clculo de las distancias, adems de entender un poco mejor su funcionamiento para proceder.
  * El algoritmo encuentra TOME LINKs para eliminar data. Es decir, busca dos datos que estn muy cercanos (similares) de diferentes clases y elimina el de la clase mayoritaria.
  * Se podra experimentar con otros algoritmos de undesampling.

```python
# Convert categorical variables to numeric
def encode_categorical_variables_pyspark(data):
  for column in data.columns:
      if isinstance(data.schema[column].dataType, StringType):
          indexer = StringIndexer(inputCol=column, outputCol=column + "_indexed")
          data = indexer.fit(data).transform(data).drop(column).withColumnRenamed(column + "_indexed", column)
  return data
```
